    1). Squeeze-and-Excitation (SE) Block
</h1>
</div>


```python
import torch.nn as nn
'''
THE SQUEEZE AND EXCITATION BLOCK
This learns to pay more attention to the most important channels from the input channels 
and puts less attention on irrelevant ones.

HOW IT WORKS
1). It finds the average of each input channel by performing global average pooling
producing a shape (B, C, 1, 1)
2). It then uses them as input into a 2-layered MLP
    Layer-1 performs dim reduction (out_channels=in_channels//reduction) and applies relu activation
    Layer-2 Projects back (out_channels=in_channels) and applies a sigmoid activation outputing channel-wise
    attention scores
3). Finally the attention scores are multiplied by their respective channels (original feature maps)       
'''
class SqueezeExcitationBlock(nn.Module):
    def __init__(self, in_channels, reduction=4):
        super(SqueezeExcitationBlock, self).__init__()
        reduced_channels = in_channels // reduction # Introduce a bottleneck
        self.pool = nn.AdaptiveAvgPool2d(1) # Global Avg Pooling->outputs a single value per channel
        self.fc = nn.Sequential(
            nn.Conv2d(in_channels=in_channels, out_channels=reduced_channels, kernel_size=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(in_channels=reduced_channels, out_channels=in_channels, kernel_size=1),
            nn.Sigmoid()
        )
    def forward(self, x):
        scale = self.fc(self.pool(x))
        return x * scale     
```

    2). MBConv Block üöÄ

```python
'''
# CONCEPT EXPLANATION OF THE MOBILE INVERTED BOTTLENECK CONVOLUTION (MBConvBlock)
=================================================================================
Summary of what the MBConvBlock does: Expansion -> Depthwise conv -> Squeeze-and-Excitation -> Project -> Residual(Skip) connection

PARAMETERS:
    stride: controls downsampling the spatial resolution
    expand_ratio: Controls the width expansion factor in the bottleneck (6 for EfficientNet-B0 MBConvs)
    se_ratio: Determines how much to shrink inside SE block (normaly 0.25)

CONCEPTS:
Expansion gives the nn more room to think before condensing the features again
Depthwise Separable convolutions, there each input channel has its own filter, reducing the number of parameters, hence cheap
'''

class MBConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride, expand_ratio, se_ratio=0.25):
        super(MBConvBlock, self).__init__()
        # Add a skip connection if the shape doesn't change across the block
        self.use_res_connect = (stride == 1 and in_channels == out_channels)
        # Adjusting the number of channels relative to the input channels
        # for expand_ratio=6, and in_channels=16, hidden_dim=6x16=96
        hidden_dim = in_channels * expand_ratio
        
        # Build all layers and blocks in the layers list
        layers = []

        # Expansion layer: A 1x1 conv that expands from in_channels to hidden_dim, e.g. from 16 to 96 adding depth
        if expand_ratio != 1:
            layers.append(
                nn.Conv2d(
                    in_channels=in_channels, 
                    out_channels=hidden_dim, 
                    kernel_size=1, 
                    bias=False
                )
            )
            layers.append(nn.BatchNorm2d(hidden_dim)) # Batch Normalisation layer to stabilize training
            layers.append(nn.SiLU()) # Swish activation
            
        # Depthwise conv layer: Independent filter for each input channel
        layers.append(
            nn.Conv2d(
                in_channels=hidden_dim,
                out_channels=hidden_dim,
                kernel_size=3, 
                stride=stride,
                padding=1,
                groups=hidden_dim,
                bias=False
            )
        )
        layers.append(nn.BatchNorm2d(hidden_dim))
        layers.append(nn.SiLU())

        # Squeeze-and-Excitation: Weighs the importance of input feature maps
        layers.append(SqueezeExcitationBlock(in_channels=hidden_dim, reduction=int(1/se_ratio)))

        # Projection layer: A 1x1 conv that scales back the channels, closing the bottleneck
        layers.append(
            nn.Conv2d(
                in_channels=hidden_dim,
                out_channels=out_channels,
                kernel_size=1,
                bias=False
            )
        )
        layers.append(nn.BatchNorm2d(out_channels))

        self.block = nn.Sequential(*layers) # The complete MBConv block

    def forward(self, x):
        out = self.block(x) # expand -> depthwise -> squeeze-and-excitation -> project
        if self.use_res_connect:
            return x + out
        else:
            return out     
```

```python
'''
num_classes: Number of classes to classify
width_mult: A scaling factor that controls how wide (how many channels) the network is
'''
class EfficientNet(nn.Module):
    def __init__(self, num_classes=1000, width_mult=1.0):
        super(EfficientNet, self).__init__()

        # Stem
        out_channels = int(32 * width_mult)
        self.stem = nn.Sequential(
            nn.Conv2d(
                in_channels=3, 
                out_channels=out_channels,
                kernel_size=3,
                stride=2, # Downsample to half the resolution
                padding=1,
                bias=False
            ),
            nn.BatchNorm2d(out_channels),
            nn.SiLU()
        )

        # MBConv Blocks
        self.blocks = nn.Sequential()
        in_channels = out_channels
        block_id = 0

        base_config = [
        # expand_ratio, in_channels, out_channels, num_blocks, strides
        [1, 32, 16, 1, 1],
        [6, 16, 24, 2, 2],
        [6, 24, 40, 2, 2],
        [6, 40, 80, 3, 2],
        [6, 40, 112, 3, 1],
        [6, 112, 192, 4, 2],
        [6, 192, 320, 1, 1]
        ]

        for expand, c_in, c_out, n, stride in base_config:
            for i in range(n): # n = number of MBConv blocks in that stage
                s = stride if i == 0 else 1 # Only downsample the first block
                block = MBConvBlock(
                    in_channels=in_channels, 
                    out_channels=c_out, 
                    stride=s,
                    expand_ratio=expand
                )
                self.blocks.add_module(f"mbconv_{block_id}", block)
                in_channels = c_out
                block_id += 1
                
        # Head
        head_channels = int(1280 * width_mult)
        self.head = nn.Sequential(
            nn.Conv2d(
                in_channels=in_channels,
                out_channels=head_channels,
                kernel_size=1,
                bias=False
            ),
            nn.BatchNorm2d(head_channels),
            nn.SiLU()
        )

        # Classifier
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.classifier = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(head_channels, num_classes)
        )

    def forward(self, x):
        x = self.stem(x)
        x = self.blocks(x)
        x = self.head(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x
```
